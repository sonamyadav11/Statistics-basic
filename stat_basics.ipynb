### **1. Types of Data**
- **Qualitative Data (Categorical)**: Data that represents categories or labels.
  - Examples:
    - Nominal: Gender (Male, Female), Colors (Red, Green, Blue).
    - Ordinal: Education level (High School, Bachelor’s, Master’s).

- **Quantitative Data (Numerical)**: Data that represents numbers.
  - Examples:
    - Interval: Temperature in Celsius (e.g., 10°C, 20°C).
    - Ratio: Height (e.g., 160 cm, 180 cm).

#### **Scales of Measurement**:
- **Nominal**: Categories without any order (e.g., Eye color: Blue, Brown).
- **Ordinal**: Ordered categories (e.g., Movie ratings: 1-star, 2-star).
- **Interval**: Numerical data with no true zero (e.g., Temperature in Celsius).
- **Ratio**: Numerical data with a meaningful zero (e.g., Weight: 0 kg is meaningful).

---

### **2. Measures of Central Tendency**
- **Mean**: The average. Use when data is symmetric and free of outliers.
  - Example: Average income of employees.
- **Median**: The middle value. Use when data is skewed or has outliers.
  - Example: Median house price in a city.
- **Mode**: The most frequent value. Use for categorical data.
  - Example: Most common blood type in a population.

---

### **3. Concept of Dispersion**
- **Dispersion**: Measures how spread out the data is.
  - **Variance**: The average of squared deviations from the mean.
  - **Standard Deviation**: The square root of variance; shows spread in the same units as the data.

---

4. Box Plot**
- A box plot visualizes data distribution, showing:
  - Minimum, first quartile (Q1), median, third quartile (Q3), and maximum.
  - Outliers, which appear as individual points.

---

5. Random Sampling**
- Random sampling ensures each member of a population has an equal chance of being selected, allowing unbiased inferences about the population.

---

6. Skewness**
- Skewness measures the asymmetry of data distribution.
  - **Positive Skew**: Long tail on the right.
  - **Negative Skew**: Long tail on the left.
  - **Effect**: Affects the mean; it gets pulled towards the tail.

---

7. Interquartile Range (IQR)**
- \( IQR = Q3 - Q1 \): Measures the middle 50% of data.
- Detects outliers using:
  - Lower Bound: \( Q1 - 1.5 \times IQR \)
  - Upper Bound: \( Q3 + 1.5 \times IQR \)

---

8. Binomial Distribution**
- Used when:
  - There are a fixed number of trials.
  - Each trial has two outcomes (success/failure).
  - Probability of success is constant.

---

9. Normal Distribution**
- Properties:
  - Symmetrical bell-shaped curve.
  - Mean = Median = Mode.
  - Follows the empirical rule:
    - 68% of data within 1 standard deviation.
    - 95% within 2.
    - 99.7% within 3.

---

10. Poisson Process**
- Real-life example: Number of calls to a call center per hour.
  - If \( \lambda = 5 \) calls/hour, probability of exactly 3 calls:
    \[
    P(X = 3) = \frac{\lambda^3 e^{-\lambda}}{3!} = \frac{5^3 e^{-5}}{6} \approx 0.14
    \]

---

11. Random Variables**
- **Random Variable**: A variable whose value depends on outcomes of a random event.
  - **Discrete**: Takes specific values (e.g., number of heads in coin tosses).
  - **Continuous**: Takes any value within a range (e.g., height).

---

12. Covariance and Correlation**
- Example dataset:
  ```plaintext
  X: [1, 2, 3, 4, 5]
  Y: [2, 4, 6, 8, 10]
  ```
- The covariance formula for two variables, X and Y, is as follows:

Covariance formula: Cov(X, Y) = Σ [(Xi - Xmean) * (Yi - Ymean)] / (n - 1)


- **Correlation**:
  ρ (X,Y) = cov (X,Y) / σX.σY. =1.0
- **Interpretation**:
  - Positive correlation indicates a strong linear relationship.
